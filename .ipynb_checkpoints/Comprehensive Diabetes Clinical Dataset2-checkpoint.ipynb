```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv1D, MaxPool1D, Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
```

```python
# convert categorical columns(gender, smoking_history, location) using LabelEncoder

label_encoders = {}
for column in ['gender', 'smoking_history', 'location']:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders = le
```

```python
data.isnull().sum()
```

```python
# prepare target variable and features

X = data.drop('diabetes', axis=1)
y = data['diabetes']
```

```python
# we use (SMOTE Tomek) for Imbalance Data

from imblearn.combine import SMOTETomek
ros = SMOTETomek()
X_smt, y_smt = ros.fit_resample(X, y)
```

```python
print (X_smt.shape)
print (y_smt.shape)
```

```python
# Apply transformation

sc = StandardScaler()
X_scaled = sc.fit_transform(X_smt)
```

```python
# splitting data

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_smt, random_state=1, test_size=0.2)
```

```python
# Early Stopping to prevent overfitting

es = EarlyStopping(monitor='val_loss', patience=15, mode='auto')
```

```python
# Reduce Learning Rate when a metric has stopped improving

rlrop = ReduceLROnPlateau(monitor='val_loss', patience=15, factor=0.2)
```

```python
model = Sequential()
model.add(Conv1D(16, 3, activation='relu', input_shape=(15,1)))
model.add(Conv1D(64, 3, activation='relu'))
model.add(Conv1D(128, 3, activation='relu'))
model.add(MaxPool1D(pool_size=2, strides=2))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.10))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.10))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.15))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model

history = model.fit(X_train, y_train, epochs=500, batch_size=800, validation_data=(X_test, y_test), callbacks=[es, rlrop])
```

```python
# plot train and validation accuracy over epochs

plt.plot(history.history['accuracy'], c='g', label='Train')
plt.plot(history.history['val_accuracy'], c='b', label='Test')
plt.legend()
plt.xlabel('counts')
plt.ylabel('Accuracy')
plt.show()
```

# KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
ks = 14
mean_acc = np.zeros((ks - 1))
std_acc = np.zeros((ks - 1))

for i in range(1, ks):
    knn_test = KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train)
    knn_test_pred = knn_test.predict(X_test)
    mean_acc[i-1] = accuracy_score(y_pred=knn_test_pred, y_true=y_test)
    std_acc[i-1] = np.std(knn_test_pred == y_test) / np.sqrt(y_test.shape[0])

print ("The Best Accuracy is: ", mean_acc.max(), " K: ", mean_acc.argmax()+1)
```

```python
# make model by KNN Classification

knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
```

```python
# predict model KNN

knn_predict = knn.predict(X_test)
```

```python
print ("Accuracy: %.4f" % accuracy_score(y_pred=knn_predict, y_true=y_test))
```

```python
knn_class = classification_report(y_pred=knn_predict , y_true=y_test)
print (knn_class)
```
